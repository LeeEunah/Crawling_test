```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

os.chdir(r'C:\Users\Administrator\Downloads\6주차_11일차,12일차_자연어처리\Week6 자연어처리 Data')
review_df=pd.read_csv("thinq_review.csv")
```

```python
#word count
word_counts = review_df['comment'].apply(lambda x:len(x.split(' ')))
```

```python
word_counts.head()
```

```python
word_counts.describe()
```

```python
# 히스토그램 그리기
plt.figure(figsize=(12, 5))
plt.hist(word_counts, bins=84, alpha=0.5, color= 'r', label='word')
plt.legend()
# 그래프 제목
plt.title('Histogram of length of review')
# 그래프 x 축 라벨
plt.xlabel('Length of review')
# 그래프 y 축 라벨
plt.ylabel('Number of review')
plt.show()
```

```python
sns.distplot(word_counts,kde=False)
```

```python
plt.figure(figsize=(12, 5))
ax = sns.distplot(word_counts,kde=False,bins=84,color='r')
ax.set_xlabel('Length of review')
ax.set_ylabel('Number of review')
ax.set_title('Histogram of length of review')
ax.legend(labels=['word'])
plt.show()
```

```python
plt.boxplot(word_counts,
             labels=['counts'],
             showmeans=True)
```

```python
sns.boxplot(word_counts)
```

```python
from wordcloud import WordCloud
font_path='c:/Windows/Fonts/'
cloud = WordCloud(font_path=font_path+'HMKMMAG.ttf',width=800, 
                  height=600, collocations = False).generate(" ".join(review_df['comment']))
plt.figure(figsize=(20, 15))
plt.imshow(cloud)
plt.axis('off')
```

```python
cloud = WordCloud(font_path=font_path+'HMKMMAG.ttf',width=800, 
                  height=600, background_color='white',collocations = False).generate(" ".join(review_df['comment']))
plt.figure(figsize=(20, 15))
plt.imshow(cloud)
plt.axis('off')
```

```python
from PIL import Image
import numpy as np

im = Image.open('heart.jpg')

mask_arr = np.array(im)
cloud = WordCloud(font_path=font_path+'HMKMMAG.ttf', mask = mask_arr,colormap='Set2',
                  width=800, 
                  height=600, background_color='white',collocations = False).generate(" ".join(review_df['comment']))
plt.figure(figsize=(20, 15))
plt.imshow(cloud)
plt.axis('off')
```
