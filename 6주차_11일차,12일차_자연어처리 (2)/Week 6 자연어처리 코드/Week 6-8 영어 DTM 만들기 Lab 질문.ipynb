2. comment_n 컬럼값을 리스트로 변환하고 다시 하나의 문자열로 만들어서 comment_n_str 컬럼을 추가하시오.

```python
review_df['comment_n_str']=review_df['comment_n'].apply(lambda x: " ".join(eval(x)))
review_df
```

3. commnet_n_str 컬럼을 사용하여 CountVectorizer를 이용하여 DTM를 생성하시오.

```python
vector = CountVectorizer()
count_dtm=vector.fit_transform(review_df['comment_n_str']).toarray() 
print(count_dtm.shape)
print(vector.vocabulary_)
```

4. commnet_n_str 컬럼을 사용하여 TfidfVectorizer를 이용하여 DTM를 생성하시오.

```python
tfidfv = TfidfVectorizer().fit(review_df['comment_n_str'])
tfidv_dtm=tfidfv.transform(review_df['comment_n_str']).toarray()
print(tfidv_dtm.shape)
print(tfidfv.vocabulary_)
```
