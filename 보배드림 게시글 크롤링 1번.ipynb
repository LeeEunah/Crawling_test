```python
import requests
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import time
from time import sleep
import random
from tqdm.auto import tqdm, trange
import pandas as pd
import numpy as np

chrome_driver = './chromedriver' # 파일 확장자 이름 미표기
headers = [{'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/90.0.4430.212 Safari/537.36'},]
```

```python
# driver = webdriver.Chrome(chrome_driver)
SEARCH_KEYWORD = input("원하는 키워드 입력 바람: ").replace(' ', '+')
# 스크래핑 할 URL 세팅 (+검색 키워드)
SEARCH_URL = "https://www.bobaedream.co.kr/list.php?code=freeb&s_cate=&maker_no=&model_no=&or_gu=10&or_se=desc&s_selday=&pagescale=30&info3=&noticeShow=&bestCode=&bestDays=&bestbbs=&vdate=&ndate=&nmdate=&s_select=Subject&s_key=" + SEARCH_KEYWORD

article_urls = []
pages = 10 ### 필요시 추가

for i in range(pages):
    print(i,'번째 페이지 시작')
    url = SEARCH_URL + "&page=" + str(i+1)
    response = requests.get(url,headers = headers[0])
    soup = BeautifulSoup(response.text, 'html.parser')
    article_list = soup.select("tbody > tr")
    for article in article_list:        
        try:
            art_url = 'https://www.bobaedream.co.kr' + str(article.select('a')[0].attrs['href'])
            article_urls.append(art_url)
        except:
            pass
uurl = pd.DataFrame(article_urls)        
print('페이지 수집 완료(1페이지당 30게시글 있음)')
```

```python
import random

num=0
title_list=[]
content_list=[]
date_list=[]
comment_content_list = []
comment_user_list = []
comment_date_list = []

for url in uurl.iloc[:,0][5:]:
    t = np.nan
    c = np.nan
    date = np.nan
    comment_content = np.nan
    comment_user = np.nan
    comment_date = np.nan
    print(num,'번째 글 시작')
    response = requests.get(url,headers = headers[0])
    time.sleep(random.random() + 0.3)
    soup = BeautifulSoup(response.content.decode('utf-8', 'replace'), 'html.parser')
    
    try :
        #제목
        t = soup.select("#print_area > div.writerProfile > dl > dt")[0].get_text()
        title_list.append(t)
        time.sleep(random.random() + 0.3)

        #내용
        article_contents = soup.select('#print_area > div.content02 > div')
        c= article_contents[0].get_text().replace("\r", "").replace("\n", "") 
        content_list.append(c)
        time.sleep(0.3)
        
        #날짜
        date = soup.select("#print_area > div.writerProfile > dl > dd.proflieInfo > ul > li:nth-child(2) > span.proCont")[0].get_text()
        date_list.append(date)
        time.sleep(0.3)
        
        #댓글
        try:
            cnt = 1
            while True:
                comment_user = soup.select("#cmt_reply > li:nth-child({0}) > dl > dt > span.name".format(cnt))[0].get_text()
                comment_date = soup.select("#cmt_reply > li:nth-child({0}) > dl > dt > span.date".format(cnt))[0].get_text()
                comment_content = soup.select("#cmt_reply > li:nth-child({0}) > dl > dd".format(cnt))[0].get_text().replace("\r", "").replace("\n", "") 
                
                comment_user_list.append(comment_user)
                comment_date_list.append(comment_date)
                comment_content_list.append(comment_content)
                cnt += 1
        except:
            pass
    except:
        t= 'nan'
        c= 'nan'
        d= 'nan'
        title_list.append(t)
        content_list.append(c)
        date_list.append(d)
        
    num=num+1
```

```python
tt = pd.DataFrame(title_list)
cc = pd.DataFrame(content_list)
dd = pd.DataFrame(date_list)
df = pd.concat([tt,cc,dd],axis=1)
df.columns = ['제목','내용','날짜']

# 제목 전처리
df['제목'] = df.apply(lambda x: x['제목'].split('[')[0], axis=1)
df['년'] = df.apply(lambda x: x['날짜'].split('.')[0], axis=1)
df['월'] = df.apply(lambda x: x['날짜'].split('.')[1], axis=1)
df['일'] = df.apply(lambda x: x['날짜'].split('.')[2], axis=1)

comment = {'내용': comment_content_list,
            '유저': comment_user_list,
            '날짜': comment_date_list}
comment_df = pd.DataFrame(comment)
```
