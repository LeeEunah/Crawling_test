```python
import requests
from urllib import request
from bs4 import BeautifulSoup
import time
import pandas as pd
import numpy as np
BASE_URL = "https://gall.dcinside.com/board/lists"
ARTICLE_BASE_URL = "https://gall.dcinside.com"
headers = [{'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/90.0.4430.212 Safari/537.36'},]
```

```python
article_urls = []
pages = 5 # demo: 5000, original: 50000
for i in range(pages):
    print(i,'번째 페이지 시작')
    url = "https://gall.dcinside.com/board/lists/?id=appliance&page=" + str(i+1)
    response = requests.get(url,headers = headers[0])
    soup = BeautifulSoup(response.text, 'html.parser')
    article_list = soup.select("tbody > tr")
    for article in article_list:        
        try:
            art_url = 'https://gall.dcinside.com' + str(article.select('a')[0].attrs['href'])
            # trash url 제거
            if ("addc.dcinside.com" in art_url) or (art_url == "https://gall.dcinside.comjavascript:;") :
                art_url = np.nan
            article_urls.append(art_url)
        except:
            pass
uurl = pd.DataFrame(article_urls)        
uurl = uurl.dropna() # 결측값 삭제
```

```python
# url 파일 저장
uurl.to_csv('url.csv')
```

```python
import random
num=0
title_list=[]
content_list=[]
for url in uurl.iloc[1:,0]: # 수정
    t = np.nan
    c = np.nan
    print(num,'번째 글 시작')
    try:
        response = requests.get(url,headers = headers[0])
        time.sleep(random.random() + 0.3)
        soup = BeautifulSoup(response.text, 'html.parser')
    #     soup = BeautifulSoup(response.content.decode('utf-8', 'replace'), 'html.parser')
    except: 
        pass
    
    try :
        #제목
        t = soup.select("h3.title.ub-word > span.title_subject")[0].get_text()
        title_list.append(t)
        time.sleep(random.random() + 0.3)
        #내용
        article_contents = soup.select('div.write_div')
        c= article_contents[0].get_text()
        content_list.append(c)
        time.sleep(0.3)
        
    except:
        print('except')
        t= 'nan'
        c= 'nan'
        title_list.append(t)
        content_list.append(c)
        
    print('제목:',t)
#     print('내용:',c)
    num=num+1
```

```python
tt = pd.DataFrame(title_list)
cc = pd.DataFrame(content_list)
df = pd.concat([tt,cc],axis=1)
df.to_csv('tc_v1.csv',encoding='utf-8')
```
